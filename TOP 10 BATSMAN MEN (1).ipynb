```python
#importing the requried Libraries
from bs4 import BeautifulSoup
import requests
```

```python
page= requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
```

```python
page.content
```

```python
soup= BeautifulSoup(page.content,'html.parser')
```

```python
batsman_name=[]
team_name=[]
batsman_rating=[]
```

```python
# let extract at the tags having the batsman-name
batsman=soup.find_all('td',class_="table-body__cell rankings-table__name name")
batsman[0:4]
```

```python
for i in batsman:
    batsman_name.append(i.get_text().replace("\n",""))
batsman_name
```

```python
# let extract at the tags having the team-name
team=soup.find_all('td',class_="table-body__cell nationality-logo rankings-table__team")
team[0:4]
```

```python
for i in team:
    team_name.append(i.get_text())
team_name
```

```python
team[0].get_text().replace("\n","")
```

```python
# let extract at the tags having the rating-name
rating=soup.find_all('td',class_="table-body__cell rating")
rating[0:4]
```

```python
for i in rating:
    batsman_rating.append(i.get_text().replace("\n",""))
batsman_rating
```

```python
print(len(batsman),
      len(team),
      len(rating))
```

```python
import pandas as pd
ODI=pd.DataFrame({})
ODI['batsman_name']=batsman
ODI['team_name']=team
ODI['batsman_rating']=rating
ODI
```
